```python
import numpy as np
g_factors = np.load('g_factors.npy')
bias_values = np.load('bias_values.npy')
tau_array = np.load('tau_values.npy')
```

```python
import matplotlib.pyplot as plt

colors = plt.get_cmap('viridis')(np.linspace(0, 1, 20))

plt.figure(figsize=(12, 8))

# Ricrea il grafico per ciascun bias_value
for i in range(20):
    row_data = tau_array[i, :]
    not_nan = ~np.isnan(row_data)
    plt.plot(g_factors[not_nan], row_data[not_nan], label=f'Bias {bias_values[i]:.3f}', color=colors[i], marker='o')

plt.title('Tau vs g_factors per Different Bias Values')
plt.xlabel('g_factors')
plt.ylabel('Tau')
plt.legend(title='Bias', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```
